#Introducció
Com en qualsevol professió tenim una terminologia pròpia i durant el capítol d'introducció s'explicaran els conceptes bàsics.

## Algoritme
Un algoritme és un conjunt ordenat d'operacions sistemàtiques que permet fer un càlcul i trobar la solució d'un problema.


##Funció##
Una funció és un bloc de codi amb un nom associat, que rep zero o més arguments d'entrada, segueix una seqüència de sentències i torna un valor i/o realitza una tasca.

##Script##
Un script és un arxiu dissenyat per ser executat. L'script por contenir programes, funcions, classes, etc.

##Mòdul##
És un script que conté coleccions de funcions, definicions i declaracions de  colecciones de `Python`.

# Tipus de dades
Els programes són formats per codi i dades, però a nivell intern de la memòria de l'ordinador no són més que una seqüència de bits.
Cada «tros» de memòria conté realment un objecte que té com a mímim els següents camps:

*   Un tipus de dada emmagatzemada.

*   Un identificador únic per distingir-lo d'altres objectes
*   Un valor consistent amb el seu tipus de dada

## Variables

**Variable.** Consta d'un espai a la memòria principal d'un ordinador i un nom simbòlic (un identificador) associat a l'espai ocupat.

La variable és la relació que hi ha entre un nom i un objecte ubicat en alguna part de la memòria.

En `Python` assignem valors a les variables seguint el format: `nom_variable = valor`

```python
x = 5
x
```

### Regles per anomenar variables

- No poden començar per nombres
- No poden començar ni contenir caràcters especials
- No poden contenir espais
- No poden anomenarse igual que les paraules reservades de Python (keywords)
- Poden començar per ' - ' (guió baix)

**Observacions**
- El llenguatge `Python`és case sensitive, és a dir diferencia entre majúscules i minúscules.
- El nom hauria de ser significatiu perquè serà més fàcil la comprensió del codi.
- En cas de tenir un noms de variables compostos, tenim 4 formes d'escriure aquests noms:
  - camelCase: `desarFitxer`
  - PascalCase: `DesarFitxer`
  - snake_case: `desar_fitxer`
  - kebab-case: `desar-fitxer`

### Operacions amb variables numériques

En el moment que desem el valor en una variable ja podem operar amb ell.

```python
x = 2
x + 5
```

Podem guardar-lo en una nova variable

```python
y = x + 5
y
```

O podem sobreescriure la mateixa variable

```python
x = x + 5
x
```

Tenim una manera alternativa per escriure `x = x + 5 ` --->  `x += 5`. A continuació es pot comprovar que el resultat és idéntic. El motiu d'aquesta nova expressió és la rapidesa.

```python
x = 3
x = x + 5
x
```

```python
x = 3
x += 5
x
```

**Observació**
- De la mateixa forma que existeix  `+=`, també tenim `-=`, `*=`, `/=`, `//=`, `%=` y `**=`. Aquestes operacions aritmétiques les anirem estudiant durant el curs.

## Comentaris

Els comentaris són la part del codi que no s'executa. Són molt necessaris per explicar blocs de codi.
Si volem fer un comentari en una línia utilitzarem `#`. En cas de que siguin més d'una línia és molt pràctic utilitzar `''' comentari  ''' `
És una bona pràctica fer comentaris per entendre el codi, però no és bo fer un ús excessiu. A l'exemple l'ús seria excessiu però s'ha considerat que és un bon exemple didàctic.

```python
'''El programa calcula l'arrel quadrada de 2
fent servir una llibreria i aquest comentati el teniu
encerclat tal com s'ha comentat'''

# Aquesta línia és un comentari i no s'executa

import math # Aquí importo una llibreria
x = 2       # Aquí declarem una variable i li assignem un valor
arrel = math.sqrt(x) # Invoco la funció arrel quadrada
# Cap comentari s'executarà
arrel
```

# Funció import

Les funcions d'un mòdul poden ser importades amb l'instrucció `import`.
Importarem el mòdul `math` que ens serà molt útil en el moment d'utilitzar funcions matemàtiques

```python
import math
```

Ara tenim carregat el mòdul `math` i podem utilitzar les seves funcions fent-ho amb la sintaxis `math.funcion()` o `math.variable`.

```python
math.sqrt(2)
```

Si volem escurçar el mom del mòdul haurem de donar la següent instrucció

```python
import math as mt
```

Una cop fet això podem invocar les funcions o variables del mòdul amb `mt`

```python
mt.sqrt(2)
```

En cas de què només volem carregar una funció o una variable del mòdul es faria de la següent manera.

```python
from math import pi
```

També podrem canviar el nom de la funció o la variable

```python
from math import pi as nombre_pi
```

I si volem més d'una funció o variabe.

```python
from math import pi, log, exp
```

Finalment, si volem carrregar tores les funcions del mòdul `math`i evitar la utilització de la sintaxis math.funcio() o math.variable, llavors la instrucció seria la següent.


```python
from math import *
```

Això últim no és gens recomanable en cas de tenir més d'un mòdul importat perquè podem tenir dos funcions o variables de diferents mòduls amb un nom idéntic i no sabrien quina hem d'utilitzar.
