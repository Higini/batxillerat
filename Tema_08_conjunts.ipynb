##Teoria de conjunts
Partirem de dos conjunts `A={1,2} i B={2,3}` per eximplificar les diferents operacions que es poden fer entre ells basats en els [Diagrames de Venn](https://es.wikipedia.org/wiki/Diagrama_de_Venn) i [Teoria de Conjunts](https://es.wikipedia.org/wiki/Teor%C3%ADa_de_conjuntos).


venn-diagrams.png




```python
A = {1, 2}

B = {2, 3}
```

###Intersecció
$A∩B$ – Elementos que están a la vez en A y en B

```python
print(A & B)
print(A.intersection(B))
```

###Unió
$A∪B$ – Elementos que están tanto en A como en B:

```python
print(A | B)
print(A.union(B))
```

###Diferència
$A−B$  – Elementos que están en A y no están en B

```python
print(A - B)

print(A.difference(B))
```

###Diferència simètrica
$A∩B$ %% – Elementos que están en A o en B pero no en ambos conjuntos

```python
print(A ^ B)

print(A.symmetric_difference(B))
```

##Inclusió
* Un conjunt B és un **subconjut** d'un altre conjunt A si tots els elements de B estan inclosos en A. un 
* Un conjunt A és un **superconjunt** d'un altre conjunt B si tots els elements de B estan inclosos en A.

```python
A = {2, 4, 6, 8, 10}
B = {4, 6, 8}
```

```python
B < A  # subconjunt
```

```python
B <= A
```

```python
A > B  # superconjunt
```

```python
B >= A
```

##Compresió de conjunts
Els conjunts, al igual que les llistes i els diccionaris, també es poden crear per compresió.
Construim un conjunt per compresió amb aquells nombres mltiples de 3 en el rang [0,20)

```python
m3 = {number for number in range(0, 20) if number % 3 == 0}

m3
```

##Conjunts inmutables
`Python`ofereix la possibilitat de crear conjunts inmutables utilitzant la funció `frozenset()`que rep qualsevol iterable com argument.

```python
marks = [1, 3, 2, 3, 1, 4, 2, 4, 5, 2, 5, 5, 3, 1, 4]

marks_levels = frozenset(marks)

marks_levels
```

```python
marks_levels.add(50)
```
