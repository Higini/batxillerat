# Conversió a tuples
Només es pot convertir a tuples aquelles estructures de dades que siguine iterables (llistes, cadena de text, diccionaris i conjunts)

```python
shopping = ['Agua', 'Aceite', 'Arroz']

tuple(shopping)
```

```python
tuple(2) # El 2 no es iterable
```

# Desempaquetat de tuples
Ens permet assignar una tupla avariables independents
[Desempaquetat](https://aprendepython.es/_images/tuple-unpacking.jpg)

```python
three_wise_men = ('Melchor', 'Gaspar', 'Baltasar')

king1, king2, king3 = three_wise_men

print(king1)
print(king2)
print(king3)
```

```python
quotient, remainder = divmod(7, 3)

print(f'El quocient és: {quotient}')
print(f'El residu és: {remainder}')

```

#Intercanvi de variables

```python
value1 = 40
value2 = 20

value1, value2 = value2, value1

print(value1)

print(value2)
```

#Desempaquetat extendido

```python
ranking = ('G', 'A', 'R', 'Y', 'W')

head, *body, tail = ranking

print(head)

print(body)

print(tail)
```

#Desempaquetat genéric

```python
oxygen = 'O2'
first, last = oxygen
first, last
```

```python
text = 'Hello, World!'
head, *body, tail = text
head, body, tail
```

```python
writer1, writer2, writer3 = ['Virginia Woolf', 'Jane Austen', 'Mary Shelley']
writer1, writer2, writer3
```

```python
text = 'Hello, World!'
word1, word2 = text.split()
word1, word2
```
