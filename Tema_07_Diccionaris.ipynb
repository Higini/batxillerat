# La funció `get()`
Per evitar els errors d'accés degut a les claus inexistents utilitzem la funció `get()`.
* Si la clau que busquem existeix, ens torna el seu valor
* Si la clau no existeix,  ens trona `None`tret del cas que li indiquem un altre per defecte.

```python
print(rae)

print(rae.get('bifronte'))

print(rae.get('programación'))

print(rae.get('programación', 'No disponible'))
```

# Combinar diccionaris
Quan combinem diccionaris hem de tenir en compte:
* Si la clau no existeix, s'afegeix amb el seu valor
* Si la clau ja existeix, s'afegeix amb el valor de l'últim diccionari

```python
rae1 = {
    'bifronte': 'De dos frentes o dos caras',
    'enjuiciar': 'Someter una cuestión a examen, discusión y juicio'
}

rae2 = {
    'anarcoide': 'Que tiende al desorden',
    'montuvio': 'Campesino de la costa',
    'enjuiciar': 'Instruir, juzgar o sentenciar una causa'
}
```

**Sense modificar** els diccionaris originals utilitzem l'operador `**`

```python
{**rae1, **rae2}
```

**Observació**

En `Python 3.9`es pot utilitzar l'operador `|` ---> `{rae1 | rae2}`

**Modificant els diccionaris originals**

```python
rae1.update(rae2)

rae1
```

#Còpies de diccionaris
Tal com va succeïr amb les llistes les modificacions d'elements d'una còpia afecta a l'original.

```python
original_rae = {
    'bifronte': 'De dos frentes o dos caras',
    'anarcoide': 'Que tiende al desorden',
    'montuvio': 'Campesino de la costa'
}

copy_rae = original_rae

original_rae['bifronte'] = 'bla bla bla'

original_rae
```

La solució és utilitza la funció `.copy()`

```python
original_rae = {
    'bifronte': 'De dos frentes o dos caras',
    'anarcoide': 'Que tiende al desorden',
    'montuvio': 'Campesino de la costa'
}

copy_rae = original_rae.copy()

original_rae['bifronte'] = 'bla bla bla'

print(original_rae)

print(copy_rae)
```

#Diccionaris per comprensió

#Objectes <\<hashables>\>
L'única restricció que han de complir les claus d'un diccionari és que han de ser <\<hashables>>. 
Un objecte és <\<hashables>> si se li pot assignar un valor <\<hash>> que no canvia en execució mai.
Per trobar el <\<hash>> d'un objecte, s'utilitza la funció `hash()`, que torna un nombre enter i es fa servir per indexar la taula <\<hash>> que es manté internament.

Els objectes han de ser **inmutables**

```python
print(hash(999))

print(hash(3.14))

print(hash('hello'))

print(hash(('a', 'b', 'c')))
```

Si l'objecte no és inmutable donarà un error

```python
hash(['a', 'b', 'c'])
```

```python
print(hash('spiderman'))

print('spiderman'.__hash__())
```
