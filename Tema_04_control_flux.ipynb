###Sentència `if anidades`

```python
temperatura = 28

if temperatura < 20:
  if temperatura < 10:
    print('Fa molt fred')
  else:
    print('Fa una mica de fred')
else:
  if temperatura < 30:
    print('Confort')
  else:
    print('Quina calor noi')

print("Continua el programa")
```

###Sentència `if-elif-else`
Sempre que tingui un else i a continuació un if ---> reduim el codi utilitzant `elif`

```python
temperatura = 28

if temperatura < 20:
  if temperatura < 10:
    print('Fa molt fred')
  else:
    print('Fa una mica de fred')
elif temperatura < 30:
    print('Confort')
else:
  print('Quina calor noi')
```

###Assignació de condicional
Són condicionals fets en una sola línia

```python
# Forma clàssica
temperatura = 35

if temperatura < 30:
  print('Azul')
else:
  print('Rojo')
```

```python
# Assignació en una única linia
temperatura = 35

print('Azul') if temperatura < 30 else print('Rojo')
```

**Observació**

Vàlid a partir de la versió `Python 3.10`

```python
command = 'Hello, World!'
match command:
  case 'Hello, World!':
     print('Hello to you too!')
  case 'Goodbye, World!':
     print('See you later')
  case other:
      print('No match found')
```

##Sentència `while`
Mentre passi això fes allò altre

**Observació**
És important donar un valor a una variable abans de comença el bucle `while`perquè canviant el valor d'aquesta variable podrem sortir del bucle

Modifiquem el programa anterior fent que després de fer-li 3 preguntes a l'usuari el programa surti fora del `while`. Com ho fem?
* Pista: Per trencar el bucle tenim la instrucció `break`

```python
vols_sortir = 'N'
nombre_preguntes = 0

while vols_sortir == 'N':

    print('Ens trobem dintre del bucle while')
    vols_sortir = (input('Vols sortir? [S/N] ')).upper()
    nombre_preguntes += 1

    if nombre_preguntes == 3:
      print('El bucle while s\'ha interromput degut a la sentència break')
      break

print('Adéu! Finalització del programa')
```

Un altre manera de fer lo mateix sense la instrucció `break`

```python
vols_sortir = 'N'
nombre_preguntes = 0

while vols_sortir == 'N' and nombre_preguntes < 3:

    print('Ens trobem dintre del bucle while')
    vols_sortir = (input('Vols sortir? [S/N] ')).upper()
    nombre_preguntes += 1

print('Adéu! Finalització del programa')
```

La sentència `while-else`
Si el bucle `while`finalitza sense arribar al `break`el flux de control passa a la sentència opcional `else`. En cas contrari, és a dir que executa `break`, no executarà la instrucció `else`

```python
vols_sortir = 'N'
nombre_preguntes = 0

while vols_sortir == 'N':

    print('Ens trobem dintre del bucle while')
    vols_sortir = (input('Vols sortir? [S/N] ')).upper()
    nombre_preguntes += 1

    if nombre_preguntes == 3:
      print('El bucle while s\'ha interromput degut a la sentència break')
      break
else:
  print('Has finalitzat el bucle while sense cap interrupció')

print('Adéu! Finalització del programa')
```

La sentència `continue`s'utiitza per saltar cap a endavant a la següent repetició.

```python
vols_sortir = 'N'
nombre_preguntes = 0

while vols_sortir == 'N':
    print('Hola!')
    vols_sortir = (input('¿Quiere salir? [S/N] ')).upper()
    if vols_sortir not in 'SN':
        vols_sortir = 'N'
        continue # Quan entri en aquest punt el progama no passarà a la instrucció nombre_preguntes += 1
    nombre_preguntes += 1
print(f'{nombre_preguntes} respuestes correctes')
print('Adéu!')
```

En el següent programa podeu veure que quan s'executa la instrucció `continue`el programa no avança més i torna a la línia del `while` sense executar la línia `nombre_preguntes  += 1`

```python
vols_sortir = 'N'
nombre_preguntes = 0

while vols_sortir == 'N':
    print('Hola!')
    vols_sortir = (input('¿Vols sortir? [S/N] ')).upper()
    continue
    nombre_preguntes += 1
print(f'{nombre_preguntes} respostes correctes')
print('Adéu!')
```

Situació de **bucle infinit**.
Podrien arribar a la situació de bucle infinit si no establim correctament la condició d'aturada, o bé tenim una variable fora de control.
**Observació**
Per avortar una situació de bucle infinit podem polsar en el teclat la combinació `Ctrl+C`. En el cas de *Google Collaboratory* heu de combinar les tecles`Ctrl+M+I` o clique sobre el botó d'execusió. L'intèrpret de `Python` mostrarà el missatge `keyboardinterrupt`

```python
n = 0

while n != 3:
  n += 2
```

Una possible solució seria

```python
n = 0

while n < 3:
  n += 2
```

A vegades un bucle infinit ens pot anar molt bé. Imaginem que un docent vol introduir les notes dels seus alumnes però sempre han d'estras dintre de l'interval [0, 10].

```python
while True:
    nota = float(input('Introdueix una nova nota: '))
    if not(0 <= nota <= 10):
        print('Nota fora de rang')
        break
    print(nota)
```

#Sentència `for`
En cas de voler iterar un nombre finit de vegades és molt útil la sentència `for`.
Sempre es pot iterar sobre els tipus de dades que siguin iterables com cadenes de text, llistes, diccionaris, tuples, conjunts,..


```python
paraula = 'Python'

for lletra in paraula:
  print(lletra)
```

Tant la comprovació de trencament d'un bucle com la continuació a la següent iteració es duen a terme de la mateixa manera que hem vist amb els bucles de tipus `while`
* Trencar una iteració `for` amb un `break`
* Salta una iteració  `for`amb un `continue`

```python
paraula = 'Python'

for lletra in paraula:
  if lletra == 'o':
    break
  print(lletra)
print('Hem finalitzat la iteracció')
```

Funció `range(start, stop, step)`
La funció `range(start, stop, step)`s'utilitza per generar una seqüència de nombres. Els arguments de la funció són:

---
`start` És opcional, valor d'inici  té el valor 0 per defecte

`stop`  És obligatori, valor de finalització - 1

`step`  És opcional, valor de salt, té el valor d'1 per defecte

---

**Observació**

`range()` torna un valor iterable



```python
for i in range(2):
  print(i)
```

```python
for i in range(1,2):
  print(i)
```

```python
for i in range(1,6,2):
  print(i)
```

```python
for i in range(2, -1, -1):
  print(i)
```

Utilització del guió baix `_` en el bucle `for`
A vegades no es necessita utilitzar una variable que prengui els valors del rang, sino que únicament volem repetir una acció un nombre determinat de vegades.
Amb el guió baix `_`donem la idea de què no estem utilitzant la variable de forma explícita.


```python
for i in range(3):
    print('Repeteix això 3 cops!')
```

#Bucles imbricats ("anidados")
L'imbricament ("anidamiento") és una tècnica per la que incloem diferents encapsulaments de sentències, unes dintre de les altres, amb un major nivell de profunditat.

```python
'''
  Taules de multiplicar de l'1 i del 2
'''
for i in range(1, 3):
    for j in range(1, 10):
        resultat = i * j
        print(f'{i} * {j} = {resultat}')
    print(20*'-')
```
