##Altres formes de concatenar strings
||
-|-|
`str()`: pot concatenar strings i variables de qualsevol tipus dintre d'un `print()`
.`format()`: Concatena strings dintre d'un `print()`
`f'{}'`: Concatena strings

* El m√®tode `format()` indica amb claus, `{}`, el lloc on volem situar el resultat de les variables. A continuaci√≥ dintre dels par√©ntesis del m√®tode `.format()` s'indiquen les variables amb el seu respectiu ordre.

Donar format a altres bases: binari, octal i hexadecimal

Donar format a valors flotants

```python
pi = 3.14159265
print(f'pi:f') # 6 decimals per defecte
print(f'{pi:.3f}') # 3 decimals --> .3f
print(f'{pi:10f}') # 8 espais inclosos els dels d√≠gits del nombre pi(8)
print(f'{pi:10.3f}')
print(f'{pi:010.3f}')
print(f'{pi:e}')
```

##Trossejar cadenes (slice)##
* `[:]` Extreu la seq√º√®ncia sencera del principi fins al final.
* `[start:]` Extreu comen√ßant pel final de la cadena.
* `[:end]` Extreu des del comen√ßament de la cadena fins end menys 1.
* `[start:end]` Extreu a partir d'start fins end menys 1.
* `[start:end:step]`Extreu a partir d'start fins end menys 1 fent salts de tamany step.

```python
frase = 'Hola companys'
print(frase[:])
print(frase[5:])
print(frase[:4])
print(frase[5:13])
print(frase[::2])
```

##Codi Unicode
`Python` treballa per defecte amb [Unicode](https://unicode-table.com/en/1F680/)
* La funci√≥ `chr()`permet representar un car√†cter a partir del seu codi
* La funci√≥ `ord()`permet obtenir el codi (decimal) d'un car√†cter a partir de la seva representaci√≥
* El modificador `\N` permet representar un car√†cter a partir del seu nom

**Observaci√≥**

Per passar d'hexadecimal a decimal `Python`afegeix `0x`davant del nombre hexadecimal


```python
codi_coet = 0x1F680
imatge_coet = chr(codi_coet)
print(imatge_coet)
```

```python
codi_decimal = ord(imatge_coet)
codi_hexadecimal = hex(codi_decimal)
print(codi_hexadecimal)
```
