# Strings en `Python`

## Variable string

**String.** És una cadena ordenada de caracters.

Una variable de tipus string és aquella que guarda un string. La manera de declarar una variable de tipus string, `str` en `Python` és envoltar el contingut el contenido entre cometes dobles `" "`, o bé entre cometes simples `' ' `.

```python
s1 = "Això és un sting entre cometes dobles"
type(s1)
```

```python
s2 = 'Això és un sting entre cometes dobles'
type(s2)
```

##Concatenació d'Strings##
La concatenació és una operació que uneix dos o més strings en un únic string suma dels altres dos.

En `Python`, per concatenar dos variables de tipus string fem servir la funció `+`.

```python
s1 = 'Hola, '
s2 = 'Esteve'
s1 + s2
```

**Observacions**
- Hem de dexar un espai en blanc al final de l'string s1 o al principi de l'string s2 per tal de no tenir enganxats els dos strings.
- El nombre d'espais en blanc es conserven

```python
s1 = 'Hola,'
s2 = '            Esteve'
s1 + s2
```

##Repetir strings##
En `Python`, per repetir una variable de tipus string utilitzem la funció `*`. L'ordre dels factors no  altera el producte i és equivalent a escriure `num_repeticions * variable_str` com `variable_str * num__repeticions`.

```python
'Papa, Hem arribat? ' * 5
```

##Altres formes de concatenar strings
||
-|-|
`str()`: pot concatenar strings i variables de qualsevol tipus dintre d'un `print()`
.`format()`: Concatena strings dintre d'un `print()`
`f'{}'`: Concatena strings

* El mètode `format()` indica amb claus, `{}`, el lloc on volem situar el resultat de les variables. A continuació dintre dels paréntesis del mètode `.format()` s'indiquen les variables amb el seu respectiu ordre.
* La sintaxis per formatejar cadenes literals dels "f-strings" és la següent f"Aquesta és una f-string {nom_var} y {nom_var}"

```python
# Concatenar amb str()
nom = "Toni"
edat = 45
print('El ' + nom + ' té ' + str(edat) + ' anys')
```

```python
# mètode format
nom = 'Xavi'
edat = 54
print('El {} té {} anys'.format(nom, edat))
```

```python
# f-string
print(f"Hola, em dic {'Pep'} i tinc {42} anys")
```

Formatejant cadenes

```python
alçada = 1234
f'{alçada:10d}' # 10 espais incloent-hi els que corresponen als dígits 1234
```

```python
f'{alçada:010d}'
```

Donar format a altres bases: binari, octal i hexadecimal

```python
valor = 0b111 # passar de codi binari a decimal amb '0b' davant del codi binari
f'{valor}'
```

```python
f'{valor:b}' # De decimal a binari
```

```python
valor = 0o111  # Passar de codi octal a decimal --> Afegir '0o' davant del valor
f'{valor}'
```

```python
f'{valor:o}' # De decimal a octal
```

```python
valor = 0x1a # Passar d'hexadecimal a decimal
f'{valor}'
```

```python
f'{valor:x}'
```

Donar format a valors flotants

```python
pi = 3.14159265
print(f'pi:f') # 6 decimals per defecte
print(f'{pi:.3f}') # 3 decimals --> .3f
print(f'{pi:10f}') # 8 espais inclosos els dels dígits del nombre pi(8)
print(f'{pi:10.3f}')
print(f'{pi:010.3f}')
print(f'{pi:e}')
```

##Trossejar cadenes (slice)##
* `[:]` Extreu la seqüència sencera del principi fins al final.
* `[start:]` Extreu començant pel final de la cadena.
* `[:end]` Extreu des del començament de la cadena fins end menys 1.
* `[start:end]` Extreu a partir d'start fins end menys 1.
* `[start:end:step]`Extreu a partir d'start fins end menys 1 fent salts de tamany step.

```python
frase = 'Hola companys'
print(frase[:])
print(frase[5:])
print(frase[:4])
print(frase[5:13])
print(frase[::2])
```

##Codi Unicode
`Python` treballa per defecte amb [Unicode](https://unicode-table.com/en/1F680/)
* La funció `chr()`permet representar un caràcter a partir del seu codi
* La funció `ord()`permet obtenir el codi (decimal) d'un caràcter a partir de la seva representació
* El modificador `\N` permet representar un caràcter a partir del seu nom

**Observació**

Per passar d'hexadecimal a decimal `Python`afegeix `0x`davant del nombre hexadecimal


```python
codi_coet = 0x1F680
imatge_coet = chr(codi_coet)
print(imatge_coet)
```

```python
codi_decimal = ord(imatge_coet)
codi_hexadecimal = hex(codi_decimal)
print(codi_hexadecimal)
```
